```python
zip_neighbor_dict = {"60621": "Englewood", 
                    "60636": "Englewood", 
                    "60619": "Chatham", 
                    "60620": "Chatham",
                    "60623": "North Lawndale",
                    "60624": "East Garfield Park",
                    "60647": "Humboldt Park",
                    "60651": "Humboldt Park"}

merged_df["neighborhood"] = merged_df["zip_clean"].map(zip_neighbor_dict)
```

```python
merged_df.head()
```

```python
# MERGING GDF ON ZIP
merged_df = merged_df.rename(columns={"zip_clean": "zip"})
merged_df['zip'] = merged_df['zip'].astype(str)
gdf['zip'] = gdf['zip'].astype(str)

merged_gdf_zip = gdf.join(merged_df.set_index('zip'), on='zip', how='inner')
```

```python
merged_gdf_zip.head()
```

```python

merged_gdf_zip['# Student Enrollment'] = (
    pd.to_numeric(merged_gdf_zip['# Student Enrollment']
                  .astype(str)
                  .str.replace(',', '', regex=True)
                  .str.strip(),
                  errors='coerce')
)

avg_enrollment_by_zip = (
    merged_gdf_zip.groupby('zip', as_index=False)['# Student Enrollment']
      .mean()
      .rename(columns={'# Student Enrollment': 'avg_enrollment'})
)

merged_gdf_zip = merged_gdf_zip.merge(avg_enrollment_by_zip, on='zip', how='left')
```

```python
import matplotlib.pyplot as plt
import matplotlib.patheffects as pe 
merged_gdf_zip.plot(column='avg_enrollment', legend=True)
```

```python
# Plot your map
fig, ax = plt.subplots(figsize=(10, 8))
merged_gdf_zip.plot(column='avg_enrollment', legend=True, ax=ax, edgecolor='black')

# Add neighborhood names at the centroid of each polygon
for idx, row in merged_gdf_zip.iterrows():
    if row['geometry'].is_empty:
        continue
    x, y = row['geometry'].centroid.x, row['geometry'].centroid.y
    name = str(row['neighborhood'])  # 👈 change to your actual column name
    ax.text(
        x, y, name,
        fontsize=7,
        ha='center', va='center',
        color='black',
        path_effects=[pe.withStroke(linewidth=2, foreground="white")]  # outline for contrast
    )

# Optional: tidy up plot
ax.set_axis_off()
plt.title("Average Enrollment by Neighborhood", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
import matplotlib.patheffects as pe

# Read the city boundary shapefile (or GeoJSON)
chicago = gpd.read_file("data_sources/shapefiles_geolocations/Neighborhoods_2012b_20251108/geo_export_d6da9fc2-1417-41bc-96a9-e15b09615172.shp")  # or .geojson, etc.

# Plot
fig, ax = plt.subplots(figsize=(10, 8))
chicago.plot(ax=ax, color='white', edgecolor='black')  # base layer
merged_gdf_zip.plot(column='avg_enrollment', legend=True, ax=ax, edgecolor='gray')

# Add neighborhood labels
for idx, row in merged_gdf_zip.iterrows():
    if row['geometry'].is_empty:
        continue
    x, y = row['geometry'].representative_point().x, row['geometry'].representative_point().y
    ax.text(x, y, str(row['neighborhood']),
            fontsize=7, ha='center', va='center',
            color='black', path_effects=[pe.withStroke(linewidth=2, foreground="white")])

ax.set_axis_off()
plt.title("Average Enrollment by Neighborhood (Chicago)", fontsize=14)
plt.tight_layout()
plt.show()
```
